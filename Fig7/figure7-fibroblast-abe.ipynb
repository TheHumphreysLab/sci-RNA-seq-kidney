```python
import numpy as np
import pandas as pd
import scanpy as sc

sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
from matplotlib import rcParams
import matplotlib.font_manager
rcParams['font.sans-serif']=['Arial']
sc.settings.set_figure_params(dpi=100, facecolor='white',fontsize=12)
from matplotlib import rcParams

adata=sc.read('read_anndata.h5ad')
```

```python
sc.pl.umap(adata,color='fib_type2',size=20,title='Fibroblast Subclustering',alpha=0.6)
```

```python
sc.settings.set_figure_params(dpi=100, facecolor='white',fontsize=12)

cell_order=[ 'JGA','Fib-1', 'Fib-2', 'Fib-3','Myo-4','Myo-1', 'Myo-2', 'Myo-3']
marker_genes=['Ren1','Flt1','Fhl2', 'Itih5','Adamtsl1','Spon1','Prickle1',
              'Il34','Il31ra','Col1a1','Acta2']

sc.pl.DotPlot(adata, marker_genes, groupby='fib_type2',
             categories_order=cell_order,standard_scale='var')\
.style(cmap="Oranges",largest_dot=400,dot_edge_lw=0,size_exponent=1.6)\
.legend(width=1.5,colorbar_title='Mean expression',size_title='Fraction of cells').show()
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 2.5,2
rcParams['figure.facecolor'] = 'black'

sc.pl.umap(adata, color='sample3_corrected',title="",size=5,alpha=0.6,
           palette=['black','black','#f0ecb9'],frameon=False,legend_loc=False,save='f5_UUO.png')#UUO
sc.pl.umap(adata, color='sample3_corrected',title="",size=5,alpha=0.6,
           palette=['black','#f0ecb9','black'],frameon=False,legend_loc=False,save='f5_IRI.png')#IRI
sc.pl.umap(adata, color='sample3_corrected',title="",size=5,alpha=0.6,
           palette=['#f0ecb9','black','black'],frameon=False,legend_loc=False,save='f5_Health.png')#Healt
```
