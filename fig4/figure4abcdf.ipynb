```python
import numpy as np
import pandas as pd
import scanpy as sc

sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
from matplotlib import rcParams
import matplotlib.font_manager
rcParams['font.sans-serif']=['Arial']
sc.settings.set_figure_params(dpi=100, facecolor='white',fontsize=12)
from matplotlib import rcParams

adata=sc.read('read_LoH_file.h5ad')
##Anndata for clustering of all LoH cells
```

```python
adata2=sc.read('read_distal_file.h5ad')
##Anndata for clustering of all DCT/CNT/PC cells
```

```python
sc.pl.umap(adata,color='loh_type',size=8,frameon=False,title='',save='f4_loh.png',alpha=0.6)
```

```python
marker_genes=['Clcnka','Sgcz','Pax2','Slc4a11','Cdh13','Slc14a2',
              'Slc12a1','Umod',
             'Kctd1','Lcn2']

sc.pl.DotPlot(adata, marker_genes, groupby='loh_type',standard_scale='var')\
.style(cmap="Reds",largest_dot=400,dot_edge_lw=0,size_exponent=1.1,dot_max=0.8).swap_axes()\
.legend(width=1.8,colorbar_title='Mean expression',size_title='Fraction of cells').show()
```
